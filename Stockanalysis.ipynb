```python
import yfinance as yf
import pandas as pd
from scipy.stats import pearsonr

# List of stock symbols for major US petroleum companies
petroleum_stocks = ['XOM', 'CVX', 'COP', 'BP', 'MRO', 'PSX', 'VLO']

# List of stock symbols for major US arms companies
arms_stocks = ['LMT', 'NOC', 'RTX', 'GD', 'BA']

# Define periods of US interventions in the Middle East, including Afghanistan
invasion_periods = [
    ('2001-10-07', '2021-08-30'),  # Afghanistan War 2001-2021
    ('2003-03-01', '2004-03-01'),  # Iraq Invasion 2003
    ('1990-08-01', '1991-08-01'),  # Gulf War 1990-1991
    ('2011-03-19', '2012-03-19'),  # Libya Intervention 2011-2012
    ('2014-09-01', '2017-12-31')   # Syrian Civil War intervention 2014-2017
]

# Download stock data for the specified companies
all_stocks = petroleum_stocks + arms_stocks
data = yf.download(all_stocks, start="1990-01-01", end="2024-01-01")['Adj Close']

# Calculate daily average stock prices for petroleum and arms companies
data['Petroleum_Avg'] = data[petroleum_stocks].mean(axis=1)
data['Arms_Avg'] = data[arms_stocks].mean(axis=1)

# Filter data for each invasion period and perform correlation analysis
correlations = {}
for start_date, end_date in invasion_periods:
    period_data = data[(data.index >= start_date) & (data.index <= end_date)]
    petroleum_avg = period_data['Petroleum_Avg']
    arms_avg = period_data['Arms_Avg']

    # Calculate Pearson correlation
    correlation, _ = pearsonr(petroleum_avg, arms_avg)
    correlations[f"{start_date} to {end_date}"] = correlation

# Display correlation results for each invasion period
for period, corr_value in correlations.items():
    print(f"Correlation between petroleum and arms stock averages during {period}: {corr_value:.4f}")
```

```python
import yfinance as yf
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# List of stock symbols for major US petroleum companies
petroleum_stocks = ['XOM', 'CVX', 'COP', 'BP', 'MRO', 'PSX', 'VLO']

# List of stock symbols for major US arms companies
arms_stocks = ['LMT', 'NOC', 'RTX', 'GD', 'BA']

# Download stock data for the specified companies
all_stocks = petroleum_stocks + arms_stocks
data = yf.download(all_stocks, start="1950-01-01", end="2024-01-01")['Adj Close']

# Calculate daily average stock prices for petroleum and arms companies
data['Petroleum_Avg'] = data[petroleum_stocks].mean(axis=1)
data['Arms_Avg'] = data[arms_stocks].mean(axis=1)

# Calculate rolling correlation between Petroleum_Avg and Arms_Avg with a 90-day window
rolling_corr = data['Petroleum_Avg'].rolling(window=90).corr(data['Arms_Avg'])

# Filter periods where the correlation is greater than or equal to 0.75
high_correlation_periods = rolling_corr[rolling_corr >= 0.75]

# Display periods with high correlation
print("Periods with rolling correlation >= 0.75 between petroleum and arms stock averages:")
for date, corr_value in high_correlation_periods.items():
    print(f"Date: {date.strftime('%Y-%m-%d')}, Correlation: {corr_value:.4f}")

# Plotting the rolling correlation
plt.figure(figsize=(12, 6))
plt.plot(rolling_corr.index, rolling_corr, label='90-Day Rolling Correlation', color='skyblue')
plt.axhline(0.75, color='red', linestyle='--', label='Threshold (0.75)')
plt.title('90-Day Rolling Correlation between Petroleum and Arms Stock Averages')
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.legend()
plt.tight_layout()
plt.show()
```
